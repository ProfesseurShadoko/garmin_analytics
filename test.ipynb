```python
from fitparse import FitFile
```

```python

fitfile = FitFile(filepath)

message_types = set()

for msg in fitfile.get_messages():
    message_types.add(msg.name)

print(sorted(message_types))

for msg in fitfile.get_messages():
    skip = ["record", "gps_metadata", "event", "lap", "device_info"]
    if msg.name in skip:
        continue
    if msg.name.startswith("unknown"):
        continue

    print(f"\nMessage type: {msg.name}")
    
    for field in msg:
        print(f"  {field.name}: {field.value}")
```
